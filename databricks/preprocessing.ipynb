Data Transformation

```python
from pyspark.sql.functions import *
```

```python
from pyspark.sql.types import StringType
```

```python
df = df.withColumn("Cylinders", col("Cylinders").cast(StringType()))
```

```python
df.display(5)
```

```python
from pyspark.sql.functions import mean


```

```python
# Calculer la moyenne de la colonne 'Weight'
mean_weight = df.select(mean('Weight_')).collect()[0][0]

```

```python
mean_weight = float(mean_weight)
```

```python
# Remplir les valeurs nulles dans la colonne 'Weight' avec la moyenne
df = df.fillna({'Weight_': mean_weight})
```

```python
# Remplir toutes les valeurs nulles avec 0 dans le DataFrame
df = df.fillna(0)
```

```python
from pyspark.sql.functions import count, when, col

# Compter les valeurs nulles dans chaque colonne du DataFrame
df.select([count(when(col(c).isNull(), 1)).alias(c) for c in df.columns]).show()
```

```python
# Supprimer les doublons sur toutes les colonnes
df = df.dropDuplicates()
```

```python
from pyspark.sql.functions import col

# Filtering rows where EngineSize is less than or equal to 5
df = df.filter(~(col("EngineSize") > 5))

# Display the result
df.show()
```

```python
from pyspark.sql.functions import col

# Ajouter 5 à chaque valeur de la colonne 'Wheelbase'
df = df.withColumn("Wheelbase", col("Wheelbase") + 5)

# Afficher le DataFrame résultant

```

```python
from pyspark.sql.functions import when, col

# Remplacer "Sedan" par "XUV" dans la colonne "Model"
df = df.withColumn("Model", when(col("Model") == "Sedan", "XUV").otherwise(col("Model")))

```

```python
from pyspark.sql.functions import year, month, dayofmonth, col

# Créer les colonnes 'year', 'month', 'day' à partir de 'date_id'
df = df.withColumn("year", year(col("date_id"))) \
         .withColumn("month", month(col("date_id"))) \
         .withColumn("day", dayofmonth(col("date_id")))


```

```python
df.write.format("parquet").mode("append").option("path","abfss://silver@carsmalyanadatalake.dfs.core.windows.net/cars_sales"
).save()
```
